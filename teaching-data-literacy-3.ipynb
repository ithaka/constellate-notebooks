## Get Started with GitHub

1. Go to [GitHub](https://github.com) and choose "Sign up".
2. Enter an email and password. (Note, you can skip the personalization questions.)
3. Verify your email.

### Fork a repository
We need to make a copy of the repository we want to base our class on. For example, you could fork any of the TAP Institute 2023 repositories:

* [Constellate Notebooks](https://github.com/ithaka/constellate-notebooks/)- Includes the 2023 TAP Institute classes in Python Basics, Pandas Basics, and Pandas Intermediate

* [Tools for taming text](https://github.com/elliewix/tapi-taming-text-2023)- Learn how to work with regular expressions and XPath

On the repository page, choose "Fork" to make your own copy of the repository in your GitHub account. It is recommended to give your forked copy a different name from the original repository so you don't get them confused.

### Create a Personal Access Token (PAT)
In order to push our changes from the Constellate lab, we need a special kind of password called a Personal Access Token. Here is how we will create it:

1. Choose the "Settings" menu under your icon in the far upper-right hand corner under the user icon.

![Selecting the user icon](https://ithaka-labs.s3.amazonaws.com/static-files/images/tdm/tdmdocs/click-on-user-icon.png)

![The settings button in the GitHub interface](https://ithaka-labs.s3.amazonaws.com/static-files/images/tdm/tdmdocs/choose-settings.png)

2. Scroll all the way to the bottom of the lefthand column to "Developer settings."

![Developer settings at the bottom](https://ithaka-labs.s3.amazonaws.com/static-files/images/tdm/tdmdocs/developer-settings.png)

3. Choose "Personal Access Tokens," then "Tokens (classic)," and finally "Generate a personal access token."

![Generate the personal access token](https://ithaka-labs.s3.amazonaws.com/static-files/images/tdm/tdmdocs/generate-pat.png)

4. Create a note that helps you understand what the token is for, for example "Constellate Lab Access." 

![Example settings for the PAT](https://ithaka-labs.s3.amazonaws.com/static-files/images/tdm/tdmdocs/pat-settings.png)

5. Choose how long you want the token to last. You can choose for the token to never expire.
6. Click the repo flag, scroll to the bottom, and choose "Generate token."
7. Copy your personal access token and store it somewhere secure, such as a password manager. For the short term, you could store it in a text document. Once you close the tab, you will not be able to see the token anymore! If you forget or lose the token, you should delete it in GitHub and generate a new token.

![An example PAT](https://ithaka-labs.s3.amazonaws.com/static-files/images/tdm/tdmdocs/your-pat.png)


## The Jupyter Lab `git` Extension

Start your Jupyter Lab environment and move into the root directory.

[Click the folders to move to root](https://ithaka-labs.s3.amazonaws.com/static-files/images/tdm/tdmdocs/move-to-root.png)

### Clone your repository

On the lefthand column, choose the git extension. Choose "Clone a Repository". Then enter the GitHub URL of your project you would like to clone into your lab environment. Remember to use the link to *your* copy of the repository (not the link to the original repository you forked from). It will look like this:

`https://github.com/your-username/your-repository-name/`

![choose to clone the repo and fill in the information](https://ithaka-labs.s3.amazonaws.com/static-files/images/tdm/tdmdocs/clone-your-repo.png)

### Push changes to your GitHub repository
There are three requirements before changes can be pushed to back to GitHub. Any files to be pushed need to be:

* Tracked (This only applies to files that git has not seen before.)
* Staged (This applies to any files you are ready to push. Usually any you have changed.)
* Committed (This creates a message to explain how and why the file was changed.)

1. Make a change to an existing file in the lab. Save those changes.
2. Check the git extension and the file will appear under "Changed". Click plus to stage the file. 

![Staging the file by clicking plus](https://ithaka-labs.s3.amazonaws.com/static-files/images/tdm/tdmdocs/stage-the-change.png)

3. When the file is under "Staged," write a comment in the Summary field and optionally a description. The summary is required before the file can be committed, so the commit button at the bottom will be gray until a comment is written. Click "Commit" when your comment is finished.

![Writing a commit message and click commit](https://ithaka-labs.s3.amazonaws.com/static-files/images/tdm/tdmdocs/commit-the-change.png)

4. At the top of the git extension, an orange circle will appear next to the cloud with an up arrow. Click this button to push your local changes to GitHub.

![Click the cloud with an up arrow to push](https://ithaka-labs.s3.amazonaws.com/static-files/images/tdm/tdmdocs/push-the-change.png)

### Untracked files
When a new file appears, it may show us an "Untracked" file. Click the + to track the file in the repository. If you would like to always ignore the file, such as a system file like `README-checkpoint.md` then right-click on the file and choose "Ignore this file (add to .gitignore)". This will permanently ignore the file so long as it stays in a list of files within a .gitignore file. Files that start with a period are invisible, but you can see this file either through the command line or on the GitHub page for the repository. The only files *required* to be in the repository are Jupyter notebook files, but it is also a good practice to include a README.md file that explains the repository for anyone discovering it on GitHub.


![Right click on an untracked file to add it to your .gitignore file](https://ithaka-labs.s3.amazonaws.com/static-files/images/tdm/tdmdocs/ignore-a-file.png)
